```python
# Import dependencies.
import matplotlib.pyplot as plt

#importing pyplot library from matplotlib
```

```python
# data in array 
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# subplots to show the object-oriented interface

fig, ax = plt.subplots()

# returns tuple containing figure and axes objects 
# fig references the figure, ax for axis of figure to help annotate


ax.plot(x_axis, y_axis)

# one liner
```

```python
# Create the plot with ax.plt()

fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)

# two step process
```

```python
plt.plot(x_axis,y_axis)

plt.show()

# actual plot instead of jupyter outputs
# better for python
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# styling with markers, colors, linewidth, labels

# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=4, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
# Change the size of the figure in pixels. Add this in the subplots() function.

ax.plot(x_axis, y_axis, label='Boston', marker="D", color="purple", linewidth=1)
# Add a label that will be added to the legend.

ax.set_xlim(0, 11)
# Sets the min and max range of the x-axis.

ax.set_ylim(0, 50)
# Sets the min and max range of the y-axis.

ax.set_xlabel('Date')
# Add a label to the x-axis.

ax.set_ylabel('Fares ($)')
# Add a label to the y-axis.

ax.set_title("PyBer Fare by Month")
# Add a title.

ax.legend()
# Add a legend.

ax.grid()

```

```python
# # scatter plot: visually depicts a positive, negative, or no-correlation relationship between two factors

# # one way with plt.plot(x, y 'o')
# plt.plot(x_axis, y_axis, 'o')

# # other 
# plt.scatter(x_axis, y_axis)

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.subplots(figsize=(20, 10))

plt.scatter(y_axis, x_axis, color="red", label="Chicago")

plt.gca().invert_yaxis()

plt.title("PyBer Fare by Month")

plt.xlabel("Fare ($)")

plt.ylabel("Month")

plt.legend()

plt.xlim(0, 50)

# plt.figure(20, 10)

# plt.subplots(figsize=(20, 10))

```

```python
# # bubble charts:  presenting financial, population, and weather data 
# # adds a third and fourth factor to convey more information
# # adds "size" as a factor, after the x_axis, y_axis 
# # adds "color" as a factor, when dataset uses the same axis

# plt.scatter(x_axis, y_axis, s=y_axis)

# # s denotes size, corresponds to the decimal point values of y_axis

y_axis_larger = []
for d in y_axis:
    y_axis_larger.append(d*3)

y_axis_larger

plt.scatter(x_axis, y_axis, s=y_axis_larger)

# # tricky, still use y_axis as y axis, but make it larger size by using dataset
# # mulitplied by 3 to make y_axis larger

# # list comprehension: optimized for Python to spot a predictable pattern during looping
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

# # cleaner
```

```python
# object-oritented approach

# # scatter
# fig, ax = plt.subplots()
# ax.scatter(x_axis, y_axis)

# # bubble
# fig, ax = plt.subplots()
# ax.scatter(x_axis, y_axis, s=y_axis)

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
# ax.scatter(x_axis, y_axis, s=y_axis)
ax.scatter(
    y_axis, x_axis,
    color="skyblue",
    label="Boston",
    s=[d*5 for d in y_axis],
    alpha=0.8, 
    edgecolor="black",
    linewidth=2)
ax.invert_yaxis()
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare $")
ax.set_ylabel("Month")
ax.set_xlim(0, 50)
```

```python
# percentages of categories as wedges of a pie
# One limitation only represent one dataset or category

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# pie(values, labels="labels")

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# explode_values= (x_axis[6]=0.2)
# standout category, 0 touchs the center, 1 is outside the diameter

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(10, 10))
# figsize needs to before setting up chart

plt.pie(y_axis,
        labels=x_axis,
       explode=explode_values,
       autopct='%.1f%%',
       colors=colors)
plt.show()
```

```python
# o-o approach

fig, ax = plt.subplots(figsize=(8,8))

explode_valuesn = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
       labels=x_axis, 
      colors=colors,
      autopct='%.1f%%',
      explode=explode_valuesn,
      shadow=True,
      startangle=90,
      counterclock=False,
      )

plt.show()

# difference between the charts
# pie: visually see the contribution of the months, with july and march being tge the biggest
# line: shows the same with line, peak in october
# v, h bars: same
# scatter: not very much
# bubble scatter: inc in size with monthhs, a 3rd data set for size might make it better
```

```python
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt
import statistics 
# stats up 

```

```python
# Get the standard deviation of the values in the y-axis.

stdev = statistics.stdev(y_axis)
stdev
```

```python
# visually show the variability of plotted data
# error bars show significance of data

plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# capsize adds a 'cap' to the error values
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# o-o approach

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
import numpy as np

# major lines (ticks) flow together with digestable gaps

plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# minor lines
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
